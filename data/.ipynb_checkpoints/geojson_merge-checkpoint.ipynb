```python
#Dependencies
import pandas as pd
import os
import json
```

```python
#load countries.geojson
with open('countries.geojson') as countriesData:
    countries_import = json.load(countriesData)
```

```python
#drill down in geojson to properties section; assign properties section to variable props
props = []

for item in countries_import["features"]:
    props.append(item["properties"])

props
```

```python
#put props into a dataframe
countries = pd.DataFrame(props)

countries.count()
```

```python
#read tourism data into a pandas dataframe called data_import
data_import = pd.read_csv('Data_Extract_From_World_Development_Indicators.csv')
```

```python
#place columns needed into a new df called data
data = data_import.loc[:,['Series Name','Series Code','Country Code','2016 [YR2016]']]
```

```python
#filter data for series about international tourism expenditures
data_expend = data.loc[data['Series Code']=='ST.INT.XPND.CD']

data_expend.head()
```

```python
#merge countries data with international tourism expenditures and assign to a new df called countries_merge
countries_merge = countries.merge(data_expend,how='left', left_on='ISO_A3',right_on='Country Code').drop(columns=['Series Name', 'Series Code','Country Code'])

countries_merge = countries_merge.rename(columns={'2016 [YR2016]':'SNT_INT_XPND_CD'})
countries_merge.head()
```

```python
#filter data for series about international arrivals
data_arrive = data.loc[data['Series Code']=='ST.INT.ARVL']

print(data_arrive.head())

#merge countries_merge data with international arrivals
countries_merge = countries_merge.merge(data_arrive,how='left', left_on='ISO_A3',right_on='Country Code').drop(columns=['Series Name', 'Series Code','Country Code'])
countries_merge = countries_merge.rename(columns={'2016 [YR2016]':'SNT_INT_ARVL'})
countries_merge.head()
```

```python
#filter data for series about international departures
data_depart = data.loc[data['Series Code']=='ST.INT.DPRT']

print(data_depart.head())

#merge countries_merge data with international departures
countries_merge = countries_merge.merge(data_depart,how='left', left_on='ISO_A3',right_on='Country Code').drop(columns=['Series Name', 'Series Code','Country Code'])
countries_merge = countries_merge.rename(columns={'2016 [YR2016]':'SNT_INT_DPRT'})
countries_merge.head()
```

```python
#filter data for series about international receipts
data_rcpt_cd = data.loc[data['Series Code']=='ST.INT.RCPT.CD']

print(data_rcpt_cd.head())

#merge countries_merge data with international receipts
countries_merge = countries_merge.merge(data_rcpt_cd,how='left', left_on='ISO_A3',right_on='Country Code').drop(columns=['Series Name', 'Series Code','Country Code'])
countries_merge = countries_merge.rename(columns={'2016 [YR2016]':'SNT_INT_RCPT_CD'})
countries_merge.head()
```

```python
#replace all '..' and nan with ''

countries_merge = countries_merge.replace('..','')
countries_merge = countries_merge.fillna('')

countries_merge.head()
```

```python
countries_merge.count()
```

```python
#move countries_merge df into a dictionary

countries_dict = countries_merge.to_dict(orient='records')

countries_dict
```

```python
#replace countries_merge data with the properties section of the geojson
for index,item in enumerate(countries_import["features"]):
    item['properties'] = countries_dict[index]
```

```python
countries_import["features"][1]
```

```python
#write file out to a countries_data geojson
with open('countries_data.geojson', 'w') as outfile:
    json.dump(countries_import, outfile)
```
